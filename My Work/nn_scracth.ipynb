```python
x1, x2 = 3, 4
y = 22

w1, w2, b = np.random.rand(3)	# setting random parameters
lr, epochs = 0.008, 20			# setting hyper parameters
losses = []

# running epochs
for _ in range(epochs):
	_y = (w1 * x1) + (w2 * x2) + (b) 	# getting the prediction
	loss = (y - _y) ** 2		# getting mean squared error
	losses.append(loss)

	# caculating derivatives
	d_w1 = 2 * (_y - y) * (x1)
	d_w2 = 2 * (_y - y) * (x2)
	d_b = 2 * (_y - y)		

	# updating the parameters
	w1 -= d_w1 * lr 
	w2 -= d_w2 * lr 
	b -= d_b * lr

	# print(f"%.4f"%loss, end="\t")

# print(w1, w2, b)
plt.scatter(range(epochs//2), losses[:epochs//2])
print("Finla Loss:", loss)
```
